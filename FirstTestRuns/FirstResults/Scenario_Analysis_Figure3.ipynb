```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
from matplotlib.lines import Line2D # For custom legend

# --- 1. Define Scenarios and Extract Steady States ---
# Assumes all 8 'out_...' variables exist.
scenarios = {
    'Cluster 1': out_baseline,
    'Bottom Up': out_bottom_up,
    'Top Down': out_hom,
    'Both low': out_both,
    'Baseline (GGE low)': out_baseline_gge,
    'Bottom Up (GGE low)': out_bottom_up_gge,
    'Top Down (GGE low)': out_hom_gge,
    'Cluster 2': out_cluster2
}

plot_data = [] # A list to hold our tidy data dictionaries

# Loop through each scenario, extract data, and format it
for scenario_name, dataset in scenarios.items():
    
    # Calculate the mean of the last 1000 days
    steady_state = dataset['Phytoplankton__biomass'].isel(time=slice(-1000, None)).mean(dim='time')
    
    # Get the size class coordinate names
    size_class_coord = steady_state.coords[steady_state.dims[0]] 
    
    # Add one entry for each size class
    for size_class in size_class_coord.values:
        plot_data.append({
            'scenario': scenario_name,
            'size_class': size_class,
            'biomass': steady_state.sel({size_class_coord.name: size_class}).item()
        })

# Convert the list of dictionaries into a pandas DataFrame
df = pd.DataFrame(plot_data)
print("\n--- Tidy DataFrame for Plotting ---")
print(df)

# --- 2. Calculate Totals for Overlay ---
df_total = df.groupby('scenario', sort=False)['biomass'].sum().reset_index()
df_total = df_total.rename(columns={'biomass': 'total_biomass'})
print("\n--- DataFrame for Totals ---")
print(df_total)
scenario_order = df['scenario'].unique() # Get the exact order from the dictionary


# --- 3. Plot the Data ---
plt.figure(figsize=(16, 6)) # Widened figure

# --- Plot the size class dots ---
ax = sns.stripplot(
    data=df,
    x='scenario',
    y='biomass',
    hue='size_class',
    order=scenario_order, # Use the defined order
    palette='tab10',
    dodge=True,
    s=12,
    alpha=0.9,
    linewidth=1,
    edgecolor='black'
)

# --- Plot the total sum as a vertical line marker ---
sns.pointplot(
    data=df_total,
    x='scenario',
    y='total_biomass',
    order=scenario_order, # Use the defined order
    ax=ax,
    marker='|',
    markersize=20,
    markeredgewidth=2.5,
    color='black',
    linestyles='',
    label='Total Biomass (Sum)'
)

# --- 4. Final Touches ---
ax.set_title("Model Scenario Comparison: Steady-State Phytoplankton Biomass", fontsize=16)
ax.set_xlabel("Model Scenario", fontsize=12)
ax.set_ylabel("Mean Biomass (Last 1000 Days)", fontsize=12)
ax.set_ylim(bottom=0)

# <-- CORRECTED: Removed 'ha="right"'
ax.tick_params(axis='x', rotation=30, labelsize=10) 

plt.grid(axis='y', linestyle='--', alpha=0.7)

# --- Custom Legend ---
handles, labels = ax.get_legend_handles_labels()
num_size_classes = len(df['size_class'].unique())
dot_handles = handles[:num_size_classes]
dot_labels = labels[:num_size_classes]
total_line_handle = Line2D([0], [0], marker='|', color='black', label='Total Biomass (Sum)',
                           markersize=20, markeredgewidth=2.5, linestyle='None')
ax.legend(
    handles=dot_handles + [total_line_handle],
    labels=dot_labels + ['Total Biomass (Sum)'],
    title='Biomass',
    bbox_to_anchor=(1.01, 1), # Adjusted position slightly
    loc='upper left'
)

plt.tight_layout()
plt.savefig('scenario_comparison_plot_8_scenarios.png', dpi=300)
plt.show()
```
